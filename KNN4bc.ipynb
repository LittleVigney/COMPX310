1. import library.

```python
import sklearn
import seaborn
import pandas
import numpy
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

2. Load data from csv file.

```python
data = pandas.read_csv("breast-cancer-wisconsin.csv")
# print(data.dtypes)
```

3. Replace "#NUM!" with the mean value of all nuclei.

```python
data.replace("#NUM!", numpy.nan, inplace=True)

print("Number of NaN value in data:")
print(data.isnull().sum())

data["nuclei"] = pandas.to_numeric(data["nuclei"], errors="coerce")
data["nuclei"] = data["nuclei"].fillna(data["nuclei"].mean())

print("Number of NaN value in data:")
print(data.isnull().sum())

# # another solution: delete the data point with invalid value
# data.replace("#NUM!", pandas.NA, inplace=True)
# data = data.dropna(X_training, )
```

4. Plot source data and print some information about data.

```python
seaborn.pairplot(data)

print("First five rows of data:")
print(data.head())
print("Shape of data:")
print(data.shape)
```

5. Split test set and training set.

```python
feature_name = ["id", "thickness", "size", "shape", "adhesion", "single", "nuclei", "chromatin", "nucleoli", "mitosis"]
label_name = ["class"]
X = data[feature_name]
y = data[label_name]

# print("First rows of X:")
# print(X.head())
# print("Shape of X:")
# print(X.shape)
# print("First rows of y:")
# print(y.head())
# print("Shape of y:")
# print(y.shape)

# The valid range of random_state is [0, 4294967295], so take last four digits of student ID as input.
X_training, X_test, y_training, y_test = train_test_split(X, y, test_size=0.2, random_state=6327)

type(X_training)
print("Shape of training set of X:")
print(X_training.shape)
print("Shape of training set of y:")
print(y_training.shape)
print("Shape of test set of X:")
print(X_test.shape)
print("Shape of test set of y:")
print(y_test.shape)
```

6. Train a kNN classifier and test it on test set.

```python
kNN_model = KNeighborsClassifier(n_neighbors=2)

kNN_model.fit(X_training, y_training.values.ravel())

y_hat = kNN_model.predict(X_test)

print("Predict results y_hat is:")
print(y_hat)
```

7. Print confusion matrix and classification report.

```python
print("Confusion matrix:")
print(confusion_matrix(y_pred=y_hat, y_true=y_test))
print("Classifier report:")
print(classification_report(y_pred=y_hat, y_true=y_test))
```

8. Train model which k from 1 to 100. And plot accuracy-k diagram.

```python
accuracy_lst = []

for k in range(1, 101):
    kNN_model = KNeighborsClassifier(n_neighbors=k)
    kNN_model.fit(X_training, y_training.values.ravel())
    y_hat = kNN_model.predict(X_test)
    accuracy_lst.append(accuracy_score(y_pred=y_hat, y_true=y_test))

k_lst = [kv for kv in range(1, 101)]

plt.scatter(k_lst, accuracy_lst)
plt.xlabel("k value")
plt.ylabel("accuracy")
plt.title("k vs accuracy diagram")
```
